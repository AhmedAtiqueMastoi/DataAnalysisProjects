### Step 4: Data Cleaning

#### Dropping duplicates:

```python
#Remove Duplicate Rows
data = data.drop_duplicates()
```

#### Remove NaN values from the dataset:

```python
#Drop rows with missing values:
data = data.dropna()
```

### Step 5: Data Visualization

#### Correlation Map for Numeric Variables:

```python
# Select numeric columns
numeric_data = data.select_dtypes(include=['int64', 'float64'])

# Compute the correlation matrix
corr_matrix = numeric_data.corr()

# Create a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',fmt='.2f')
plt.title("Correlation Map for Numerical Columns")
plt.show()
```

###  Visualization of the Variables

```python
# Calculate the counts of employees with and without attrition
attrition_counts = data['Attrition'].value_counts()

# Create a pie chart
plt.figure(figsize=(7, 5))
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', colors=['skyblue', 'red'])
plt.title("Employee Attrition Distribution")
plt.show()
```

##### 1.Overtime

```python
overtime_counts = data['OverTime'].value_counts()
plt.figure(figsize=(4, 6))
sns.barplot(x=overtime_counts.index, y=overtime_counts.values)
plt.xlabel('OverTime')
plt.ylabel('Number of Employees')
plt.title('Distribution of OverTime')
plt.show()
```

##### 2.MaritalStatus

```python
marital_status_counts = data['MaritalStatus'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(marital_status_counts, labels=marital_status_counts.index, autopct='%1.1f%%', startangle=140, colors=['orange', 'lightgreen', 'skyblue'])
plt.title("Distribution of Marital Status")
plt.show()
```

##### 3.Job Role

```python
JobRole_counts = data['JobRole'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(JobRole_counts, labels=JobRole_counts.index, autopct='%1.1f%%', startangle=140, colors=['orange', 'lightgreen','skyblue'])
plt.title("Distribution of Job Role")
plt.show()
```

##### 4.Gender

```python
Gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(Gender_counts, labels=Gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['orange', 'lightgreen'])
plt.title("Distribution of Gender")
plt.show()
```

##### 5.Education Field

```python
import matplotlib.pyplot as plt

# Calculate the counts for each education field
education_field_counts = data['EducationField'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(education_field_counts, labels=education_field_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Education Field")
plt.show()
```

##### 6.Department

```python
department_counts = data['Department'].value_counts()

# Create a bar chart
plt.figure(figsize=(4, 6))
plt.bar(department_counts.index, department_counts.values, color='skyblue')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Distribution of Department')
plt.show()
```

##### 7.Business Travel

```python
business_travel_counts = data['BusinessTravel'].value_counts()

# Create a bar chart
plt.figure(figsize=(5, 6))
plt.bar(business_travel_counts.index, business_travel_counts.values, color='skyblue')
plt.xlabel('Business Travel')
plt.ylabel('Number of Employees')
plt.title('Distribution of Business Travel')
plt.show()
```

## Histograms

#### Total Working Years - Histogram:

```python
plt.figure(figsize=(6, 6))
plt.hist(data['TotalWorkingYears'], bins=20, color='lightgreen', edgecolor='black')
plt.title("Distribution of Total Working Years")
plt.xlabel("Total Working Years")
plt.ylabel("Frequency")
plt.show()
```

#### Education Level - Bar Plot:

```python
plt.figure(figsize=(6, 6))
sns.countplot(data=data, x='Education')
plt.title("Education Level Distribution")
plt.xlabel("Education Level")
plt.ylabel("Frequency")
plt.show()
```

#### Number of Companies Worked - Box Plot:

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=data, y='NumCompaniesWorked')
plt.title("Distribution of Number of Companies Worked")
plt.ylabel("Number of Companies Worked")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['DistanceFromHome'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Distance from Home")
plt.xlabel("Distance from Home")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Age', y='OverTime')
plt.title("Relationship Between Overtime and Age")
plt.xlabel("Age")
plt.ylabel("Overtime")
plt.show()
```
